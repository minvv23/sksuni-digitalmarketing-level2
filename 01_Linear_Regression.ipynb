### 3. Banking Data

```python
bankingdf = pd.read_csv('./data/L5_banking.csv')
bankingdf = bankingdf.rename({'inc':'income'}, axis=1)
bankingdf.head(5)
```

```python
bankingdf.isna().any(axis=1).value_counts()
```

```python
bankingdf = bankingdf.dropna().reset_index(drop=True)
bankingdf.shape
```

```python
bankingdf = bankingdf.dropna().reset_index(drop=True)
bankingdf['age'] = (bankingdf['age']*10).astype(int)
bankingdf['income'] = bankingdf['income'].replace(
    {1:7500, 2:17500, 3:25000, 4:35000, 5:45000, 6:62500, 7:87500, 8:112500, 9:130000}).astype(int)
bankingdf.head(5)
```

```python
model = sm.OLS.from_formula("profit~online+income+age", data=bankingdf).fit(cov_type='hc3')
model.summary()
```

```python
bankingdf.groupby(['age', 'income']).mean()
```

### 4. Nonlinear Term in Regression

```python
fig = sns.lmplot(y='income', x='age', data=bankingdf, order=1)
plt.show()

fig = sns.lmplot(y='income', x='age', data=bankingdf, order=2)
plt.show()
```

```python
model = sm.OLS.from_formula("income~age+C(district)", data=bankingdf).fit(cov_type='hc3')
model.summary()
```

```python
model = sm.OLS.from_formula("income~I(age**2)+C(district)", data=bankingdf).fit(cov_type='hc3')
model.summary()
```
